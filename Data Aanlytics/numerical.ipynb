```python
#Boolean Filter
y = np.random.randint(1,100, (10,10))
print(y)
```

```python
for i in y:
    for j in i:
        if j>50:
            print (j)
```

```python
y > 50 #boolean filter
```

```python
y [y > 50] #boolean filter based data
```

```python
y [y > 150] #null array output, as condition not satisfied.
```

```python
f1 = y >10
f2 = y <50
```

```python
y[f1]
```

```python
y[f1 & f2] #AND operator
```

```python
y[f1 | f2]
```
